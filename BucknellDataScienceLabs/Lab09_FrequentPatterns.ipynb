### 20) [M] In the context of association rules, explain the difference between support, confidence, lift, leverage and conviction. 

min_support of 0.5 won't work because our largest percentage of data was 0.33. 


Support- an indication of how frequently the itemset appears in the dataset. 

Confidence- an indication of how often the rule has been found to be true. 

Lift- The ratio of the observed support to the expected support if the two variables were independent. 

Conviction- (1 - support of the dependent variable) / (1 - the confidence of the independent variable mapped onto the dependent variable)
